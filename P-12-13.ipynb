# IQR

```python
# Calculate quartiles and interquartile range (IQR)
q1 = df['F.E'].quantile(0.25)
q3 = df['F.E'].quantile(0.75)

iqr = q3 - q1

# Calculate the upper and lower bounds for outliers
lower = q1 - 1.5*iqr
higher = q3 + 1.5*iqr
```

```python
# Handling outliers using IQR
outliers = df[(df['F.E'] > higher) | (df['F.E'] < lower)]
```

```python
outliers
```

```python
higher
```

```python
# Remove outliers using IQR
df_removed = df[(df['F.E'] <= higher) & (df['F.E'] >= lower)]
print(df_removed)
```

# 3 Data Normalization using any 2 technique 1) min max and 2) decimal scaling

```python
# Min Max 
df['F.E'] = ( df['F.E'] - df['F.E'].min() ) / (df['F.E'].max() - df['F.E'].min())
df['T.E'] = ( df['T.E'] - df['T.E'].min() ) / (df['T.E'].max() - df['T.E'].min())
print(df)
```

```python
columns_to_normalize = ['S.E', 'T.E', 'B.E']

# Apply normalization to each column
for column in columns_to_normalize:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

# Display the modified DataFrame
print(df)
```

# Decimal Scaling 

```python
for column in df.columns[2:]:
    max_value = df[column].max()
    num_digits = len(str(int(max_value)))
    df[column] = df[column] / (10 ** num_digits)
```
