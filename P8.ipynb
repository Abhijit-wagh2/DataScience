```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
data = pd.read_csv('social_network_ads.csv')
```

```python
data.head()
```

```python
X = data[['Age','EstimatedSalary']]
```

```python
Y = data['Purchased']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Initialize and train the Naïve Bayes model (Gaussian Naïve Bayes in this case)
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)
```

```python
# Compute the confusion matrix
conf_matrix = confusion_matrix(Y_test, y_pred)
```

```python
# Extract TP, FP, TN, FN from the confusion matrix
TP = conf_matrix[1,1]
FP = conf_matrix[0,1]
TN = conf_matrix[0,0]
FN = conf_matrix[1,0]
```

```python
Actual \ Predicted | Negative (0) | Positive (1)
------------------------------------------------
Negative (0)       |     TN       |     FP
------------------------------------------------
Positive (1)       |     FN       |     TP
In this diagram:

TN (True Negatives): Actual negative cases correctly predicted as negative.
FP (False Positives): Actual negative cases incorrectly predicted as positive.
FN (False Negatives): Actual positive cases incorrectly predicted as negative.
TP (True Positives): Actual positive cases correctly predicted as positive.
```

```python
# Compute accuracy, error rate, precision, and recall
```

```python
accuracy = accuracy_score(Y_test, y_pred)
```

```python
error_rate = 1 - accuracy
```

```python
precision = precision_score(Y_test, y_pred)
```

```python
recall = recall_score(Y_test, y_pred)
```

```python
# Print the confusion matrix and metrics
print("Confusion Matrix:")
print(conf_matrix)
print("True Positives:", TP)
print("False Positives:", FP)
print("True Negatives:", TN)
print("False Negatives:", FN)
print("Accuracy:", accuracy)
print("Error Rate:", error_rate)
print("Precision:", precision)
print("Recall:", recall)
```
